```python
#real-time twitter stream

from tweepy.streaming import StreamListener
from tweepy import Stream

class Listener(tweepy.StreamListener):
    
    def on_connect(self):
        """Called when the connection is made"""
        print("You're connected to the streaming server.")
    
    def on_status(self, status):
        if status.retweeted_status:
            return print(status.user.name, status.text)
            

    def on_error(self, status_code): 
        if status_code == 420:
            #disconnette lo stream in caso di errore 420
            return False 
        
    def on_data(self, data):
            
        """This will be called each time we receive stream data"""
        client = MongoClient('localhost', 27017)
        # si usa database chiamato twitterdb 
        #se non dovesse esistere in MongoDB viene creato nuovo
        db = client.TwitterDB
    
        # Decode the JSON from Twitter
        datajson = json.loads(data)
            
        print("Tweet collected") 
            
        #inserisce i tweet in una collezione nel database twitterdb
        #se la collezione  nuova viene creata
        db.NBAfinalstweet.insert_one(datajson)

        
        
stream = Listener
stream = Stream(auth = api.auth, listener=stream())
stream.filter(track=['NBAplayoffs','NBA','NBAFinals','GoldenStateWarriors','TorontoRaptors'])
```

Iniziate lo stream per un po' e poi interrompete il kernel con il tasto stop (quadrato nero).
Tutti i tweet sono salvati in MongoDB nel db twitterdb nella collezione.
Per ogni stream si possono creare nuove collezioni o nuovi database.
